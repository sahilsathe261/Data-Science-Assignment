```python
df[numerical_colm].hist(figsize=(10, 10), bins=30)
plt.suptitle('Histograms of Numerical Columns')
plt.show()
```

```python
df[numerical_colm].plot(kind='box')
plt.suptitle('Boxplots of Numerical Columns')
plt.show()
```

```python
categorical_colm = df.select_dtypes(include=['object']).columns
```

```python
for col in categorical_colm:
  df[col].value_counts().plot(kind='bar', figsize=(8, 6))
  plt.title(f'Bar Chart of {col}')
  plt.xlabel(col)
  plt.ylabel('Count')
  plt.show()
```
