```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X= data.drop(columns='Type')
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
#***********************************PCA**********************************************
```

```python
from sklearn.decomposition import PCA
pca=PCA()
```

```python
X_pca=pca.fit_transform(X_scaled)
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
plt.figure(figsize=(8, 6))
plt.plot(np.cumsum(explained_variance_ratio), marker='o', linestyle='--', color='b')
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
explained_variance_ratio
```

```python
pca_5 = PCA(n_components=5)
X_pca_5 = pca_5.fit_transform(X_scaled)

X_pca_5.shape
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt

kmeans_original = KMeans(n_clusters=3, random_state=42)
clusters_original = kmeans_original.fit_predict(X_scaled)

silhouette_original = silhouette_score(X_scaled, clusters_original)
davies_bouldin_original = davies_bouldin_score(X_scaled, clusters_original)

print("Clustering on Original Data:")
print(f"Silhouette Score: {silhouette_original}")
print(f"Davies-Bouldin Score: {davies_bouldin_original}")


kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca_5)

silhouette_pca = silhouette_score(X_pca_5, clusters_pca)
davies_bouldin_pca = davies_bouldin_score(X_pca_5, clusters_pca)

print("\nClustering on PCA-Transformed Data:")
print(f"Silhouette Score: {silhouette_pca}")
print(f"Davies-Bouldin Score: {davies_bouldin_pca}")

plt.scatter(X_pca_5[:, 0], X_pca_5[:, 1], c=clusters_pca, cmap='viridis')
plt.title('Clustering on PCA-Transformed Data (First 2 Components)')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```

```python
#Conclusion
```

1. Key Findings:
Clustering on PCA-transformed data resulted in more distinct and compact clusters compared to the original data.
The first 5 principal components captured 80% of the variance, effectively simplifying the dataset without significant information loss.
PCA improved clustering performance but reduced interpretability since the components are combinations of original features.
2. Practical Implications:
PCA reduces noise and redundancy, enhancing clustering in high-dimensional datasets.
It improves computational efficiency, especially for large datasets.
The trade-off is reduced interpretability, as the clusters are harder to explain in terms of the original features.
3. Recommendations:
Use PCA when handling high-dimensional data to improve efficiency and clustering quality.
Use original data if interpretability is crucial, or when working with small, manageable datasets.
